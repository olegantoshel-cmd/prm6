```python
print("=" * 50)
print("ЗАДАНИЕ 1: СИСТЕМА ЗАКАЗОВ КАФЕ")
print("=" * 50)

menu = {}
orders = []
order_counter = 1

def add_dish():
    print("\n--- Добавление нового блюда ---")
    name = input("Название блюда: ")
    price = float(input("Цена: "))
    category = input("Категория (Основное/Первое/Закуска/Напитки): ")
    cooking_time = int(input("Время приготовления (мин): "))

    dish_id = len(menu) + 1
    menu[dish_id] = {
        "name": name,
        "price": price,
        "category": category,
        "time": cooking_time,
        "orders_count": 0
    }
    print(f"✅ Блюдо '{name}' добавлено с ID: {dish_id}")

def show_menu():
    if not menu:
        print("Меню пусто")
        return
    print("\n--- МЕНЮ ---")
    for dish_id, dish in menu.items():
        print(f"ID: {dish_id} | {dish['name']} | {dish['price']} руб. | {dish['category']}")

def create_order():
    global order_counter
    print("\n--- Создание заказа ---")
    show_menu()

    if not menu:
        return

    table = input("Номер стола: ")
    dish_input = input("Введите ID блюд через пробел: ").split()

    dish_ids = []
    for d in dish_input:
        try:
            dish_ids.append(int(d))
        except:
            pass

    order_dishes = []
    total_price = 0

    for dish_id in dish_ids:
        if dish_id in menu:
            dish = menu[dish_id]
            order_dishes.append(dish["name"])
            total_price += dish["price"]
            menu[dish_id]["orders_count"] += 1

    if order_dishes:
        order = {
            "id": order_counter,
            "table": table,
            "dishes": order_dishes,
            "price": total_price,
            "status": "готовится"
        }
        orders.append(order)
        print(f"✅ Заказ #{order_counter} создан на сумму {total_price} руб.")
        order_counter += 1
    else:
        print("❌ Заказ не создан")

def show_orders():
    if not orders:
        print("Заказов нет")
        return
    print("\n--- ВСЕ ЗАКАЗЫ ---")
    for order in orders:
        print(f"Заказ #{order['id']} | Стол {order['table']} | {order['price']} руб. | {order['status']}")
        print(f"   Блюда: {', '.join(order['dishes'])}")

def get_statistics():
    print("\n--- СТАТИСТИКА ---")

    if not menu:
        print("Нет данных")
        return

    popularity = []
    for dish_id, dish in menu.items():
        popularity.append((dish["name"], dish["orders_count"]))
    popularity.sort(key=lambda x: x[1], reverse=True)
    print(f"Популярные блюда: {popularity[:3]}")

    revenue = {}
    for order in orders:
        for dish_name in order["dishes"]:
            for dish_id, dish in menu.items():
                if dish["name"] == dish_name:
                    category = dish["category"]
                    price = dish["price"]
                    if category in revenue:
                        revenue[category] += price
                    else:
                        revenue[category] = price
                    break
    print(f"Выручка по категориям: {revenue}")

while True:
    print("\n" + "=" * 30)
    print("1. Добавить блюдо")
    print("2. Показать меню")
    print("3. Создать заказ")
    print("4. Показать заказы")
    print("5. Статистика")
    print("6. Выход")

    choice = input("Выберите действие: ")

    if choice == "1":
        add_dish()
    elif choice == "2":
        show_menu()
    elif choice == "3":
        create_order()
    elif choice == "4":
        show_orders()
    elif choice == "5":
        get_statistics()
    elif choice == "6":
        print("До свидания!")
        break
    else:
        print("Неверный выбор")
```

```python
print("=" * 50)
print("ЗАДАНИЕ 2: УЧЕТ УСПЕВАЕМОСТИ")
print("=" * 50)

students = {}
grades = []
student_counter = 1
grade_counter = 1

def add_student():
    global student_counter
    print("\n--- Добавление студента ---")
    name = input("ФИО студента: ")
    group = input("Группа: ")
    contacts = input("Контакты: ")

    student_id = student_counter
    students[student_id] = {
        "name": name,
        "group": group,
        "contacts": contacts,
        "grades": []
    }
    student_counter += 1
    print(f"✅ Студент {name} добавлен с ID: {student_id}")

def show_students():
    if not students:
        print("Студентов нет")
        return
    print("\n--- СПИСОК СТУДЕНТОВ ---")
    for student_id, student in students.items():
        print(f"ID: {student_id} | {student['name']} | {student['group']}")

def add_grade():
    global grade_counter
    print("\n--- Добавление оценки ---")
    show_students()

    if not students:
        return

    try:
        student_id = int(input("ID студента: "))
        if student_id not in students:
            print("Студент не найден")
            return

        subject = input("Предмет: ")
        grade = int(input("Оценка (2-5): "))
        if grade < 2 or grade > 5:
            print("Оценка должна быть от 2 до 5")
            return

        date = input("Дата (ГГГГ-ММ-ДД): ")

        grade_record = {
            "id": grade_counter,
            "student_id": student_id,
            "subject": subject,
            "grade": grade,
            "date": date
        }
        grades.append(grade_record)
        students[student_id]["grades"].append(grade)
        grade_counter += 1
        print(f"✅ Оценка {grade} по {subject} добавлена")
    except:
        print("Ошибка ввода")

def calculate_student_avg(student_id):
    if student_id in students and students[student_id]["grades"]:
        grades_list = students[student_id]["grades"]
        return sum(grades_list) / len(grades_list)
    return 0

def show_student_grades():
    show_students()
    if not students:
        return

    try:
        student_id = int(input("ID студента: "))
        if student_id in students:
            student = students[student_id]
            avg = calculate_student_avg(student_id)
            print(f"\n{student['name']} ({student['group']})")
            print(f"Все оценки: {student['grades']}")
            print(f"Средний балл: {avg:.2f}")

            student_grades = [g for g in grades if g["student_id"] == student_id]
            print("Детально:")
            for g in student_grades:
                print(f"  {g['subject']}: {g['grade']} ({g['date']})")
    except:
        print("Ошибка ввода")

def find_low_performance():
    try:
        threshold = float(input("Введите пороговый балл (например 3.5): "))
    except:
        threshold = 3.0

    low_students = []
    for student_id, student in students.items():
        if student["grades"]:
            avg = calculate_student_avg(student_id)
            if avg < threshold:
                low_students.append((student["name"], student["group"], avg))

    if low_students:
        print(f"\nСтуденты с баллом ниже {threshold}:")
        for name, group, avg in low_students:
            print(f"{name} ({group}) - {avg:.2f}")
    else:
        print("Таких студентов нет")

def show_rating():
    rating = []
    for student_id, student in students.items():
        if student["grades"]:
            avg = calculate_student_avg(student_id)
            rating.append((student["name"], student["group"], avg))

    rating.sort(key=lambda x: x[2], reverse=True)

    print("\n--- РЕЙТИНГ СТУДЕНТОВ ---")
    for i, (name, group, avg) in enumerate(rating, 1):
        print(f"{i}. {name} ({group}) - {avg:.2f}")

while True:
    print("\n" + "=" * 30)
    print("1. Добавить студента")
    print("2. Показать всех студентов")
    print("3. Добавить оценку")
    print("4. Показать оценки студента")
    print("5. Найти слабых студентов")
    print("6. Показать рейтинг")
    print("7. Выход")

    choice = input("Выберите действие: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        show_students()
    elif choice == "3":
        add_grade()
    elif choice == "4":
        show_student_grades()
    elif choice == "5":
        find_low_performance()
    elif choice == "6":
        show_rating()
    elif choice == "7":
        print("До свидания!")
        break
    else:
        print("Неверный выбор")
```
